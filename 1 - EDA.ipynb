```python
#headerless dataframe 
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("diabetes.csv",header=None)
df.columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI',
'DiabetesPedigreeFunction', 'Age', 'Outcome']
df.to_csv("new_diabetes.csv",index=False)
```

```python
#Counting the number of observations of the outcome varible
count=df.value_counts("Outcome")
count.plot(kind='bar')
plt.xlabel("Outcome")
plt.ylabel("counts")
plt.title("Outcomes chart")
plt.show()
```

```python
#Summary statistics
df.describe()
```

```python
#outlier detction 
import seaborn as sns
sns.boxplot(data=df[['Pregnancies','Glucose','BloodPressure','SkinThickness']])
sns.boxplot(data=df['Age'])
plt.show()
```

```python
#Checking for null values if present
df.isnull().sum()
```

```python
#to check for the correlation matrix
corr_matrix=df.corr()
sns.heatmap(corr_matrix,annot=True)
plt.title("Heatmap of the dataset")
plt.show()
```

```python
sns.pairplot(df)
plt.title("Pairplot generation of dataset")
plt.show()
```
